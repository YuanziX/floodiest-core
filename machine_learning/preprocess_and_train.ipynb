```python
import subprocess
from subprocess import PIPE
import rasterio
import json
import glob
import pandas as pd
import os
import numpy as np
import cv2
```

```python
# The local folder path where the labels are stored
path_labels = '/content/sen12flood/sen12floods_s2_labels/sen12floods_s2_labels'

# The local folder where the training images are stored
path_training_images = '/content/sen12flood/sen12floods_s2_source/sen12floods_s2_source'
```

```python
def image_label(product_id):

    rootdir = "/content/sen12flood/sen12floods_s2_labels/sen12floods_s2_labels/sen12floods_s2_labels_"
    pd = product_id.split("_")
    pd = pd[3] + "_" + pd[4] + "_" + pd[5] + "_" + pd[6]

    json_data=open(rootdir + pd +"/stac.json", "rb")
    jdata = json.load(json_data)
    flood = jdata["properties"]["FLOODING"]


    if (flood == "False"):
        image_label = 0
    else:
        image_label = 1

    return image_label
```

```python
def product_name(path):

    json_data=open(path+"/stac.json", "rb")
    jdata = json.load(json_data)

    return  jdata['id']
```

```python
def load_data():
    images = []
    labels = []

    for folder in imagelist:
        try:
            product_id = product_name(folder)
            label = image_label(product_id)
            image = cv2.imread(folder + "/stack.tif")
            images.append(image)
            labels.append(label)
        except Exception as e:
            pass


    images = np.array(images, dtype = 'float32')
    labels = np.array(labels, dtype = 'int32')

    return images, labels
```

```python
imagelist = []
rootdir = path_training_images
for file in os.listdir(rootdir):
    d = os.path.join(rootdir, file)
    if os.path.isdir(d):
        imagelist.append(d)


print(f"The number of training samples is currently = {len(imagelist)}")
```

```python
train_images, train_labels = load_data()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

import keras
from keras.models import Sequential
from keras.layers import Dense, Conv2D , MaxPool2D , Flatten , Dropout
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from keras.optimizers import Adam
from keras.utils import plot_model
import pydot

# from sklearn.metrics import classification_report,confusion_matrix

import tensorflow as tf

import cv2
import os

import numpy as np
```

```python
model = tf.keras.Sequential([
    tf.keras.Input(shape=(512, 512, 3)),
    tf.keras.layers.Conv2D(32, (3, 3), activation = 'relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Conv2D(32, (3, 3), activation = 'relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation=tf.nn.relu),
    tf.keras.layers.Dense(2, activation=tf.nn.softmax)
])
```

```python
model.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(train_images, train_labels, batch_size=16, epochs=8, validation_split = 0.2)
```
